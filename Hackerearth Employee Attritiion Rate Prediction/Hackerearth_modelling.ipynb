## Random Forest Regressor

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from sklearn import model_selection
from sklearn import metrics
from sklearn import ensemble



SEED = 42

FOLD=10
X=train.drop('Attrition_rate',1)
y=(train['Attrition_rate'])

Xt=test.copy()
test_preds = np.zeros(len(Xt))

# k-fold cross validation
scores=list()
kf = KFold(n_splits=FOLD, shuffle=True,random_state=SEED)

# enumerate splits

for train_idx, test_idx in tqdm(kf.split(X, y)):
    # get data
    X_train, y_train = X.iloc[train_idx, :], y[train_idx]
    X_test, y_test = X.iloc[test_idx, :], y[test_idx]
    test=Xt
    # fit model
    model =RandomForestRegressor(random_state=9,bootstrap=True,max_depth=5,max_features='sqrt',min_samples_leaf=4,
                                min_samples_split=2,n_estimators=300)
    model.fit(X_train,np.log1p(y_train))
    # evaluate model
    yhat = np.expm1((model.predict(X_test)))
    test_preds += np.expm1((model.predict(test)))/FOLD
    acc=100*max(0,1-np.sqrt(mean_squared_error(y_test,yhat)))
    


    
    # store score
    scores.append(acc)
    print('> ', acc)
    
# summarize model performance
# summarize model performance
mean_s= np.mean(scores)
print('Mean: %.3f' % (mean_s))
```

## Ridge Regression

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from sklearn import model_selection
from sklearn import metrics
from sklearn import ensemble



SEED = 42

FOLD=10
X=train.drop('Attrition_rate',1)
y=(train['Attrition_rate'])

Xt=test.copy()
test_preds = np.zeros(len(Xt))

# k-fold cross validation
scores=list()
kf = KFold(n_splits=FOLD, shuffle=True,random_state=SEED)

# enumerate splits

for train_idx, test_idx in tqdm(kf.split(X, y)):
    # get data
    X_train, y_train = X.iloc[train_idx, :], y[train_idx]
    X_test, y_test = X.iloc[test_idx, :], y[test_idx]
    test=Xt
    # fit model
    model =Ridge(alpha=500)
    model.fit(X_train,np.log1p(y_train))
    # evaluate model
    yhat = np.expm1((model.predict(X_test)))
    test_preds += np.expm1((model.predict(test)))/FOLD
    acc=100*max(0,1-np.sqrt(mean_squared_error(y_test,yhat)))
    


    
    # store score
    scores.append(acc)
    print('> ', acc)
    
# summarize model performance
# summarize model performance
mean_s= np.mean(scores)
print('Mean: %.3f' % (mean_s))
```
